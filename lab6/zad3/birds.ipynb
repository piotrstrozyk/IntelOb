```python
import cv2
import numpy as np
import os
from PIL import Image

folder_path = 'bird_miniatures'
def grayscale_weighted(image):
    weights = [0.299, 0.587, 0.114]
    gray = np.dot(image[..., :3], weights)
    return gray.astype(np.uint8)


def adjust_brightness_contrast(image, alpha=1.0, beta=0):
    adjusted = cv2.convertScaleAbs(image, alpha=alpha, beta=beta)
    return adjusted

image = cv2.imread('bird_miniatures/E0454_TR0000_OB0010_T01_M02.jpg')
gray = grayscale_weighted(image)
adjusted = adjust_brightness_contrast(gray, alpha=1.5, beta=30)


# threshold 
th, threshed = cv2.threshold(adjusted, 10, 255, 
       cv2.THRESH_BINARY_INV|cv2.THRESH_OTSU) 


# findcontours 
cnts = cv2.findContours(gray, cv2.RETR_LIST, 
                    cv2.CHAIN_APPROX_SIMPLE)[-2] 


# filter by area 
s1 = 3
s2 = 20
xcnts = [] 
for cnt in cnts: 
    if s1<cv2.contourArea(cnt) <s2: 
        xcnts.append(cnt) 


print("\nDots number: {}".format(len(xcnts))) 


cv2.imshow("bird", threshed)
cv2.waitKey(0)
cv2.destroyAllWindows()

# for filename in os.listdir(folder_path):

#     file_path = os.path.join(folder_path, filename)
#     # Przetwarzanie pliku - tutaj można wykonać dowolne operacje na obrazie
#     image = cv2.imread(file_path)
#     # Możesz tutaj wykonać dowolne operacje na obrazie, np. analizę, przetwarzanie itp.
#     gray = grayscale_weighted(image)
#         # Dostosowanie jasności i kontrastu
#     adjusted = adjust_brightness_contrast(gray, alpha=1.5, beta=30)
    
#     # Progowanie dwuprogowe - ustalenie progów ręcznie do separacji ptaków od nieba
#     _, binary_birds = cv2.threshold(adjusted, 10, 255, cv2.THRESH_BINARY)
#     _, binary_sky = cv2.threshold(adjusted, 200, 255, cv2.THRESH_BINARY)
    
#     # Odwrócenie maski ptaków
#     binary_birds = cv2.bitwise_not(binary_birds)
    
#     # Połączenie obu progów
#     binary_combined = cv2.bitwise_or(binary_birds, binary_sky)
    
#     # Zastosowanie maski na oryginalnym obrazie
#     result = cv2.bitwise_and(image, image, mask=binary_combined)
#     # Znalezienie konturów
#     contours, _ = cv2.findContours(binary_birds, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    
#     # Liczenie liczby konturów (czarnych plam)
#     num_spots = len(contours)
#     print(f'Znaleziono {num_spots} ptaków na obrazie {filename}')
#     # cv2.imshow(filename, result)
#     # # image.save('processed_' + filename)  # Zapisuje przetworzony obraz
#     # cv2.waitKey(0)
#     # cv2.destroyAllWindows()
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize

with open('article.txt', 'r', encoding='utf-8') as file:
    text = file.read()

sentences = sent_tokenize(text)
words = word_tokenize(text)

print('Number of sentences:', len(sentences))
print('Number of words:', len(words))
print('Number of unique words:', len(set(words)))
print(sentences[:2])
print(words[:10])
```

```python
from nltk.corpus import stopwords

print(stopwords.words('english'))
```

```python
stop_words = set(stopwords.words('english'))

# converts the words in word_tokens to lower case and then checks whether 
#they are present in stop_words or not
filtered = [w for w in words if not w.lower() in stop_words]
#with no lower case conversion
filtered = []
 
for w in words:
    if w not in stop_words:
        filtered.append(w)
 
print(filtered)
```

```python
print('Number of words:', len(filtered))
print('Number of unique words:', len(set(filtered)))
print(filtered[:100])
```

```python
from nltk.stem import WordNetLemmatizer
 
lemmatizer = WordNetLemmatizer()
 
lemmatized = [lemmatizer.lemmatize(word) for word in filtered]

print(lemmatized[:100])
print('Number of words:', len(lemmatized))
print('Number of unique words:', len(set(lemmatized)))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import pandas as pd

# Convert the lemmatized words back into a document
document = ' '.join(lemmatized)

# Create a CountVectorizer and fit it to the document
vectorizer = CountVectorizer()
X = vectorizer.fit_transform([document])

# Convert the vectorized data into a DataFrame
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Sum the counts and sort by frequency
word_counts = df.sum().sort_values(ascending=False)

# Get the 10 most common words
top_10 = word_counts[:10]

# Plot the 10 most common words
plt.figure(figsize=(10, 6))
plt.bar(top_10.index, top_10.values)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 10 most common words')
plt.show()
```

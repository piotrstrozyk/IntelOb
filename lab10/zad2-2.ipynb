```python
import matplotlib.pyplot as plt
import random
from aco import AntColony
import time

plt.style.use("dark_background")
```

```python
COORDS = (
    (20, 52),
    (43, 50),
    (20, 84),
    (70, 65),
    (29, 90),
    (87, 83),
    (73, 23),
    (10, 30),
    (65, 85),
    (90, 70),
    (50, 10),
    (40, 95),
    (25, 75),
    (80, 45),
)
```

```python
def random_coord():
    r = random.randint(0, len(COORDS))
    return r
```

```python
def plot_nodes(w=12, h=8):
    for x, y in COORDS:
        plt.plot(x, y, "g.", markersize=15)
    plt.axis("off")
    fig = plt.gcf()
    fig.set_size_inches([w, h])
```

```python
def plot_all_edges():
    paths = ((a, b) for a in COORDS for b in COORDS)

    for a, b in paths:
        plt.plot((a[0], b[0]), (a[1], b[1]))
```

```python
plot_nodes()
```

```python
start1 = time.time()
colony1 = AntColony(
    COORDS,
    ant_count=500,
    alpha=1,
    beta=2,
    pheromone_evaporation_rate=0.5,
    pheromone_constant=1000.0,
    iterations=300,
)
end1 = time.time()
print(f"Time taken for colony1: {end1 - start1}")

optimal_nodes_1 = colony1.get_path()

for i in range(len(optimal_nodes_1) - 1):
    plt.plot(
        (optimal_nodes_1[i][0], optimal_nodes_1[i + 1][0]),
        (optimal_nodes_1[i][1], optimal_nodes_1[i + 1][1]),
    )

plt.show()
```
